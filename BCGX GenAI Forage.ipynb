```python
import pandas as pd
```

```python
df = pd.read_csv('BCGX-Forage Correction.csv')
```

## Revenue Growth (%) and Net Income Growth (%)
Checking the revenue and net income growth over the last three fiscal years by grouping the respective companies together and seeing the change in percentages.

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
```

Displaying the current table with the findings

```python
df
```

## Asset Growth (%) and Liabilities Growth Growth (%)
Checking the asset and liabilities growth over the last three fiscal years by grouping the respective companies together and seeing the change in percentages.

```python
df['Asset Growth (%)'] = df.groupby(['Company'])['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby(['Company'])['Total Liabilities'].pct_change() * 100
```

Displaying the current table with the findings

```python
df
```

# Findings and Conclusions
## Revenue Growth
Looking at the revenue growth over the 3 fiscal years, Microsoft revenue growth has been decreasing by 13% for the past 2 years. Tesla had a big increase in 2023, but only had a small increase in  2024. Apple in 2024 had a small revenue gain, but had a 4% increase in 2025.

## Net Income Growth
Microsoft net income has been decreasing over the last two fiscal years. Tesla experience a decrease in net income by over 50% in 2024 after a 18% increase in 2023. Apple experienced a small decreasae of 3% in 2024, but had an increase in 19% in the following year.

## Asset Growth
Both  Microsoft and Tesla experienced an increase in the last 2 fiscal years in their assets. Tesla had a notable drop from 29% to 14% in 2024. Apple had a small 3.5% increase in 2024 and went under by 1.5% in the following year. For all 3 companies, the percentage of their asset in the most recent fiscal year didn't match the  last year.

## Liabilities  Growth
All companies experienced a decrease  in their liabilities in the latest fiscal year. Notably, Apple had a 7% decrease in liabilities. Both Microsoft and Tesla had around a 18% increase in liabilities in their last fiscal year and  around a 13% increase in the most recent.

# Trends
